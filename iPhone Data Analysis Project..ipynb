```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Data/apple_products.csv")
print(df.head())
```

```python
ram_counts = df['Ram'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(ram_counts,
        labels=ram_counts.index,
        autopct='%1.1f%%',
        startangle=90)

plt.title("iPhone RAM Distribution")
plt.show()
```

```python
df['Model Name'] = df['Product Name'].str[6:15]

top_models = df['Model Name'].value_counts().head(5)

plt.figure(figsize=(7,7))
plt.pie(top_models,
        labels=top_models.index,
        autopct='%1.1f%%',
        startangle=140)

plt.title("Top 5 iPhone Models")
plt.show()
```

```python
ram_counts = df['Ram'].value_counts()

plt.figure(figsize=(7,7))

plt.pie(ram_counts,
        labels=ram_counts.index,
        autopct='%1.1f%%',
        wedgeprops={'width':0.4})

plt.title("Donut Chart of RAM Distribution")

plt.show()
```

```python
# Extract model names
df['Model Name'] = df['Product Name'].str[6:15]

# Top 5 most common models
top_models = df['Model Name'].value_counts().head(5)

# Bar Graph
plt.figure(figsize=(8,5))

plt.bar(top_models.index, top_models.values)

plt.title("Top 5 Most Common iPhone Models")
plt.xlabel("iPhone Models")
plt.ylabel("Count")

plt.xticks(rotation=20)

plt.show()
```

```python
from wordcloud import WordCloud
```

```python
df = pd.read_csv("Data/apple_products.csv")
print(df.head())
```

```python
# Combine all product names into one text
text = " ".join(df['Product Name'].astype(str))

# Generate Word Cloud
wordcloud = WordCloud(
    width=1000,
    height=500,
    background_color='white'
).generate(text)

# Display Word Cloud
plt.figure(figsize=(12,6))

plt.imshow(wordcloud, interpolation='bilinear')

plt.axis("off")

plt.title("Word Cloud of iPhone Product Names")

plt.show()
```
